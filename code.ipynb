```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv(r"C:\AI WORKSHOP\cold drinks task\Untitled spreadsheet - Cold_Drink_Sales.csv")
```

```python
X = df[["Temperature_C"]]
y = df["Sales"]
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
import pandas as pd

p_sales = model.predict(pd.DataFrame([[40]], columns=["Temperature_C"]))[0]
print(f"{p_sales:.2f}")
```

```python
y_pred = model.predict(X)
residuals = y - y_pred
```

```python
plt.scatter(df["Temperature_C"], residuals, alpha=0.7)
plt.axhline(0, color="red", linestyle="--")
plt.xlabel("Temperature (°C)")
plt.ylabel("Residuals")
plt.title("Residuals vs Temperature")
plt.show()
```
